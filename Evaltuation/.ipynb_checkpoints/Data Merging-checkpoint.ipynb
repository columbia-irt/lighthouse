```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('close\SamsungS7 Hue Values.csv', delimiter = ';')
df2 = pd.read_csv('close\SamsungS7 Parameters.csv', delimiter = ';')

df = pd.merge(df1,df2)

df = df.set_index('TimeStamp')

df.to_csv('close\SamsungS7.csv', sep = ';')
df.head()
```
